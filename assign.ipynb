## Q 2. Given an unsorted array of integers, find the length of the longest continuous
increasing subsequence (subarray). 
Example 1:
Input: [1,3,5,4,7]
Output: 3 
Example 2:
Input: [2,2,2,2,2]
Output: 1

```python
def find_length_of_LCIS(nums):
    if not nums:
        return 0

    max_len = 1
    current_len = 1

    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_len += 1
        else:
            max_len = max(max_len, current_len)
            current_len = 1

    return max(max_len, current_len)

# Example 1
nums1 = [1, 3, 5, 4, 7]
print(find_length_of_LCIS(nums1))  # Output: 3

# Example 2
nums2 = [2, 2, 2, 2, 2]
print(find_length_of_LCIS(nums2))  # Output: 1
```

## Q 3. Given a list of non negative integers, arrange them such that they form the largest
number.
 
 Example 1:
 Input: [10,2]
 Output: &quot;210&quot;
 Example 2:
 
 Input: [3,30,34,5,9]
 Output: &quot;9534330&quot;

```python
def compare(x, y):
    # Custom comparison function
    if x + y > y + x:
        return True
    else:
        return False

def bubble_sort(arr):
    # Implement bubble sort based on custom comparison
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if not compare(arr[j], arr[j + 1]):
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def largest_number(nums):
    # Convert integers to strings for concatenation
    str_nums = [str(num) for num in nums]
    
    # Sort the array using bubble sort and custom comparison
    bubble_sort(str_nums)
    
    # Join the sorted numbers into a single string
    result = ''.join(str_nums)
    
    # Edge case: if the result is all zeros, return '0'
    if result[0] == '0':
        return '0'
    else:
        return result

# Example 1
nums1 = [10, 2]
print(largest_number(nums1))  # Output: "210"

# Example 2
nums2 = [3, 30, 34, 5, 9]
print(largest_number(nums2))  # Output: "9534330"
```

## Q 4. Store all the &quot;servlet-name&quot;, and &quot;servlet-class&quot; to a csv file from the attached
sample_json.json file using Python.

```python
import json
import csv

# Load the JSON data
with open('DT A1 sample_json (1) (1) (3).json', 'r') as file:
    data = json.load(file)

# Open a CSV file for writing
with open('output.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    
    # Write the header
    csvwriter.writerow(['servlet-name', 'servlet-class'])
    
    # Write the servlet data
    for servlet in data['web-app']['servlet']:
        csvwriter.writerow([servlet['servlet-name'], servlet['servlet-class']])

print("Data has been written to output.csv")
```
